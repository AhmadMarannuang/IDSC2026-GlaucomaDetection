```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

print("✅ Library successfully imported!")
```

```python
# Load augmented dataset
df = pd.read_csv('../glaucoma_augmented_dataset.csv')

print("Dataset loaded:")
print(f"Total data: {len(df)}")
print(f"\nLabel distribution:")
print(df['Label'].value_counts())
print(f"\nTotal unique patients: {df['Patient'].nunique()}")
```

Sesuai modul — split harus per pasien bukan per gambar untuk hindari data leakage!

```python
# Ambil unique patient IDs
patients = df['Patient'].unique()
print(f"Total pastient: {len(patients)}")

# Split pasien 80% train, 20% test
train_patients, test_patients = train_test_split(
    patients,
    test_size=0.2,
    random_state=42
)

# Split pasien train menjadi 80% train, 20% val
train_patients, val_patients = train_test_split(
    train_patients,
    test_size=0.2,
    random_state=42
)

print(f"\nNumber of train patients: {len(train_patients)}")
print(f"Number of validation patients: {len(val_patients)}")
print(f"Number of test patients: {len(test_patients)}")

# Bagi gambar berdasarkan pasien
train_df = df[df['Patient'].isin(train_patients)]
val_df = df[df['Patient'].isin(val_patients)]
test_df = df[df['Patient'].isin(test_patients)]

print(f"\nNumber of train images: {len(train_df)}")
print(f"Number of validation images: {len(val_df)}")
print(f"Number of test images: {len(test_df)}")
```

```python
# Cek distribusi label di setiap set
print("Train label distribution:")
print(train_df['Label'].value_counts())

print("\nValidation label distribution:")
print(val_df['Label'].value_counts())

print("\nTest label distribution:")
print(test_df['Label'].value_counts())
```

Kenapa Val & Test Kurang Seimbang?

Karena kita split per pasien — augmentasi hanya ada di beberapa pasien, jadi distribusinya tidak merata di val & test


Ini normal dan wajar — yang penting train seimbang karena di situlah model belajar! ✅

```python
# Reset index
train_df = train_df.reset_index(drop=True)
val_df = val_df.reset_index(drop=True)
test_df = test_df.reset_index(drop=True)

# Simpan ke CSV
train_df.to_csv('../train_dataset.csv', index=False)
val_df.to_csv('../val_dataset.csv', index=False)
test_df.to_csv('../test_dataset.csv', index=False)

print("✅ Dataset saved successfully!")
print(f"Train: {len(train_df)} images")
print(f"Val: {len(val_df)} images")
print(f"Test: {len(test_df)} images")
```
